```python
import os
import pandas as pd
import numpy as np

# Define the path to the directory
dir_path = 'archive'

# Get a list of all files in the directory
files = os.listdir(dir_path)

# Initialize an empty list to store the DataFrames
dfs = []
data = pd.DataFrame()

# Iterate over the list of files
for file in files:
    # Check if the file is a CSV file
    if file.endswith('.csv'):
        # Construct the full file path
        file_path = os.path.join(dir_path, file)
        # Load the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        # Append the DataFrame to the list
        dfs.append(df)
        df['brand'] = file.split('.')[0]
        data = pd.concat([data,df])
for dataset in dfs:

    print(f'\n')
    print(f'dataset: {dataset.columns}')
    print(f'\n')
```

# Feature scaling



```python
 # Feature scaling
''' 
Recomendado cuando los datos tienen diferentes escalas, por ejemplo, si una columna tiene valores entre 0 y 1 y otra entre 0 y 1000.
El escalado de características es un paso importante en la preparación de datos para algoritmos de aprendizaje automático que se ven afectados por la escala de las características, como la regresión lineal, la regresión logística y las redes neuronales.
Recomendado cuando la distribución de las características es asimétrica o con valores atípicos.
Recomendado cuando se utilizan algoritmos que utilizan medidas de distancia, como k-vecinos más cercanos (KNN) o k-medias.

'''
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, MaxAbsScaler, Normalizer
scaler = MaxAbsScaler()
''' 
StandardScaler: Escala las características eliminando la media y escalando a la varianza unitaria.
RobustScaler: Escala las características utilizando estadísticas robustas para evitar que los valores atípicos afecten la escala.
MinMaxScaler: Escala las características en un rango dado, generalmente entre 0 y 1.
MaxAbsScaler: Escala las características en el rango [-1, 1] dividiendo por el valor máximo absoluto.
Normalizer: Escala las características a una norma unitaria.
'''
scaler.fit(train[['mileage']])
```

```python
# Pipe lines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler, MinMaxScaler
from sklearn.pipeline import FeatureUnion
from sklearn import set_config
```

```python
# One-Hot Encoding maker, model, transmission, fuelType
one_hot_encoding = ColumnTransformer([
    (
        'one_hot_encoding', # Nombre de la transformación
        OneHotEncoder(),  # Transformador
        ['brand','transmission', 'fuelType'] # Columnas a transformar
    )
])
```

```python
one_hot_encoding.fit(train)
```

```python
#RobustScaler encoding mileage
robust_encoding = ColumnTransformer([
    (
        'robust_encoding', # Nombre de la transformación
        RobustScaler(),  # Transformador
        ['mileage'] # Columnas a transformar
    )
])
```

```python
#Impute and standar scaler 
impute_and_scale = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

standar_scaling = ColumnTransformer([
    (
        'standar_scaling', # Nombre de la transformación
        impute_and_scale,  # Transformador
        ['mpg', 'tax'] # Columnas a transformar
    )
])
```

```python
#Just pass year and engineSize
pass_through = ColumnTransformer([
    (
        'pass_through', # Nombre de la transformación
        'passthrough',  # Transformador
        ['year', 'engineSize'] # Columnas a transformar
    )
])
```

```python
#Ensambling the pipeleines
preprocessor = Pipeline([
    ('preprocessing', FeatureUnion([
        ('one_hot_encoding', one_hot_encoding),
        ('robust_encoding', robust_encoding),
        ('standar_scaling', standar_scaling),
        ('pass_through', pass_through)
    ]))
])
```

```python
set_config(display='diagram')
preprocessor
```

```python
#train_preprocessed = preprocessor.fit_transform(train)
train_preprocessed = preprocessor.transform(train)
val_preprocessed = preprocessor.transform(val)
test_preprocessed = preprocessor.transform(test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
predicting_pipe = Pipeline([
    ('feature_engineering', preprocessor),
    ('predictor', lr)
])
```

```python
set_config(display='diagram')
predicting_pipe
```

```python
_ = predicting_pipe.fit(train, train['price'])
```
